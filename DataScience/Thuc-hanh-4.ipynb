```python
#yc4.1
import pandas as pd

df_patient = pd.read_csv('Data_Patient.csv', index_col='id')
df_patient = df_patient.rename(columns={
    'feature_1': 'Age',
    'feature_2': 'Gender',
    'feature_3': 'Type',
    'feature_4': 'Blood_pressure',
    'feature_5': 'Cholesterol',
    'feature_6': 'Heartbeat',
    'feature_7': 'Thalassemia',
    'feature_8': 'Result'
})

print(df_patient)
```

```python
#yc4.2
df_male = df_patient[df_patient['Gender'] == 'Male']
df_female = df_patient[df_patient['Gender'] == 'Female']
df_no = df_patient[df_patient['Result'] == 0]
df_yes = df_patient[df_patient['Result'] == 1]

print("Danh sách bệnh nhân nam:")
print(df_male.head())
print("\nDanh sách bệnh nhân nữ:")
print(df_female.head())
print("\nDanh sách những người không bị bệnh tim:")
print(df_no.head())
print("\nDanh sách những người bị bệnh tim:")
print(df_yes.head())
```

```python
#yc4.3
df_condition1 = df_patient[(df_patient['Result'] == 1) & (df_patient['Age'] > 70)]
df_condition2 = df_patient[(df_patient['Gender'] == 'Female') & (df_patient['Blood_pressure'] == 170) & (df_patient['Result'] == 0)]
df_condition3 = df_patient[(df_patient['Type'] == 'Typical angina') & (df_patient['Gender'] == 'Male') & (df_patient['Result'] == 1)]

print("Những người mắc bệnh tim và trên 70 tuổi:")
print(df_condition1.head())
print("Người có giới tính Female, huyết áp trên 170 mmhg nhưng không bị đau tim:")
print(df_condition2.head())
print("Những người có triệu chứng đau ngực, giới tính nam và bị bệnh tim:")
print(df_condition3.head())
```

```python
#yc4.4
min_blood_pressure = df_patient['Blood_pressure'].min()
max_blood_pressure = df_patient['Blood_pressure'].max()
mean_blood_pressure = df_patient['Blood_pressure'].mean()
median_blood_pressure = df_patient['Blood_pressure'].median()
std_dev_blood_pressure = df_patient['Blood_pressure'].std()

print(f"Chỉ số huyết áp thấp nhất: {min_blood_pressure}")
print(f"Chỉ số huyết áp cao nhất: {max_blood_pressure}")
print(f"Chỉ số huyết áp trung bình: {mean_blood_pressure}")
print(f"Chỉ số huyết áp trung vị: {median_blood_pressure}")
print(f"Chỉ số độ lệch chuẩn của huyết áp: {std_dev_blood_pressure}")
print("------------------------------------------------------------")

min_heartbeat = df_patient['Heartbeat'].min()
max_heartbeat = df_patient['Heartbeat'].max()
mean_heartbeat = df_patient['Heartbeat'].mean()
median_heartbeat = df_patient['Heartbeat'].median()
std_dev_heartbeat = df_patient['Heartbeat'].std()

print(f"Chỉ số nhịp tim thấp nhất: {min_heartbeat}")
print(f"Chỉ số nhịp tim cao nhất: {max_heartbeat}")
print(f"Chỉ số nhịp tim trung bình: {mean_heartbeat}")
print(f"Chỉ số nhịp tim trung vị: {median_heartbeat}")
print(f"Chỉ số độ lệch chuẩn của nhịp tim: {std_dev_heartbeat}")
```
