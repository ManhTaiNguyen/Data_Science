```python
import numpy as np
import matplotlib.pyplot as plt

#yc1
data_diamond = np.loadtxt("Diamonds.txt", dtype=np.float64)
print(data_diamond)
print("--------------------------------------------")

#yc2
print("Kích thước biến date_diamond:", data_diamond.shape)
print("Số chiều của biến date_diamond:", data_diamond.ndim)
print("Kiểu dữ liệu của các phần tử:", data_diamond.dtype)
print("Số phần tử:", data_diamond.size)
diamond_size,diamond_price = np.hsplit(data_diamond,2)
diamond_size = np.transpose(diamond_size,(1,0))
diamond_price = np.transpose(diamond_price,(1,0))
print("Vector diamond_size:", diamond_size)
print("Vector diamond_price:", diamond_price)

#yc3
plt.scatter(diamond_size, diamond_price)
plt.title("BIỂU ĐỒ THỂ HIỆN MỐI TƯƠNG QUAN GIỮA TRỌNG LƯỢNG (CARAT) VÀ GIÁ BÁN KIM CƯƠNG($)")
plt.xlabel("Giá bán ($)")
plt.ylabel("Trọng lượng (Carat)")
plt.show()

#yc4
co = np.corrcoef(diamond_size,diamond_price)
print("Hệ só tương quan giữa trọng lượng và giá bán kim cương:", co.min())
```
