```python
import csv
import pandas as pd

df_patient = pd.read_csv("Data_Patient.csv")
file_path = "Data_Patient.csv"

with open(file_path, "r") as csv_file:
    csv_reader = csv.DictReader(csv_file)

    for row in csv_reader:
        id = row["id"]
        feature_1 = row["feature_1"]
        feature_2 = row["feature_2"]
        feature_3 = row["feature_3"]
        feature_4 = row["feature_4"]
        feature_5 = row["feature_5"]
        feature_6 = row["feature_6"]
        feature_7 = row["feature_7"]
        feature_8 = row["feature_8"]

        print(f"Mã của bệnh nhân: {id}")
        print(f"Tuổi của bệnh nhân: {feature_1}")
        print(f"Giới tính của bệnh nhân: {feature_2}")
        print(f"Triệu chứng: {feature_3}")
        print(f"Huyết áp: {feature_4}")
        print(f"Chỉ số Cholesterol của bệnh nhân: {feature_5}")
        print(f"Thông số nhịp tim của bệnh nhân: {feature_6}")
        print(f"Chỉ số Thalassemia của bệnh nhân: {feature_7}")
        print(f"Kết luận: {feature_8}")
```

```python
#yc3.1
df_patient = df_patient.rename(columns={
    "feature_1": "Age",
    "feature_2": "Gender",
    "feature_3": "Type",
    "feature_4": "Blood_pressure",
    "feature_5": "Cholesterol",
    "feature_6": "Heartbeat",
    "feature_7": "Thalassemia",
    "feature_8": "Result"
})

print("Thông tin tổng quan của DataFrame:")
print(df_patient.info())
print("\nThông tin của 10 bệnh nhân đầu tiên:")
print(df_patient.head(10))
print("\nThông tin của 5 bệnh nhân cuối cùng:")
print(df_patient.tail(5))
print("\nDataFrame sau khi đặt lại tên cột:")
print(df_patient.head())
```

```python
#yc3.2
df_Age = pd.read_csv('Data_Patient.csv')

min_age = df_Age['feature_1'].min()
max_age = df_Age['feature_1'].max()
print(f"Tuổi của bệnh nhân trẻ nhất: {min_age}")
print(f"Tuổi của bệnh nhân già nhất: {max_age}")

mean_cholesterol = df_Age['feature_5'].mean()
std_cholesterol = df_Age['feature_5'].std()
print(f"Cholesterol trung bình của bệnh nhân: {mean_cholesterol}")
print(f"Độ lệch chuẩn của Cholesterol: {std_cholesterol}")

male_count = df_Age[df_Age['feature_2'] == 'Male'].shape[0]
print(f"Số lượng bệnh nhân giới tính nam: {male_count}")

unique_types = df_Age['feature_3'].nunique()
most_common_type = df_Age['feature_3'].mode()[0]
most_common_type_count = df_Age['feature_3'].value_counts().max()
print(f"Số lượng giá trị khác nhau của thuộc tính Type: {unique_types}")
print(f"Giá trị xuất hiện nhiều nhất của thuộc tính Type: {most_common_type}")
print(f"Số lần xuất hiện của giá trị {most_common_type}: {most_common_type_count}")
```

```python
#yc3.3
df = pd.read_csv('Data_Patient.csv')

missing_data = df.isna().sum()
print("Số lượng dữ liệu bị thiếu cho từng cột:")
print(missing_data)
```

```python
gender_mapping = {'Male': 0, 'Female': 1}
df['feature_3'] = df['feature_3'].replace(gender_mapping)

result_mapping = {0: 'No', 1: 'Yes'}
df['feature_8'] = df['feature_8'].replace(result_mapping)

patient_05_index = 'Patient_05'
new_thalassemia_value = 4.0
df.loc[df.index == patient_05_index, 'Thalassemia'] = new_thalassemia_value

print(df)
```
